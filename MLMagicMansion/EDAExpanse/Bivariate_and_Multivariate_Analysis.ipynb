# Numerical and Categorical -- BarPlot , BoxPlot , DistPlot

```python
titanic
```

```python
sns.barplot(x = "Pclass" , y = "Age" , data=titanic , hue = 'Sex')
```

```python
sns.boxplot(x = 'Sex' , y = 'Age' , data = titanic , hue = 'Survived')
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

Using sns.distplot in Seaborn allows you to visualize the distribution of data. However, note that sns.distplot has been deprecated and replaced by sns.histplot and sns.kdeplot in newer versions of Seaborn. For distribution plots without histograms, sns.kdeplot is recommended.

```python
sns.histplot(titanic[titanic['Survived'] == 0]['Age'], kde=True, color='red', label='Did Not Survive', bins=30)

sns.histplot(titanic[titanic['Survived'] == 1]['Age'], kde=True, color='blue', label='Survived', bins=30)
```

```python
sns.kdeplot(data=titanic[titanic['Survived'] == 0]['Age'], label='Did Not Survive', shade=True)

sns.kdeplot(data=titanic[titanic['Survived'] == 1]['Age'], label='Survived', shade=True)
```
